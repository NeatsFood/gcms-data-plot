```python
import numpy as np
import pandas as pd
import itertools
from bokeh.plotting import figure, show, output_file
from bokeh.models import ColumnDataSource, LabelSet
from bokeh.palettes import Dark2_5 as palette

#Replace this selection with automatic code to look for file paths and break up a csv automatically
#Code - Matlabers - Reminder for manu to push that
df = pd.read_csv("data.csv")
h_1 = df[2:13110]
h_2 = df[13113:26221]
h_3 = df[26224:39332]
h_4 = df[39335:52443]
h_5 = df[52446:65554]
h_6 = df[65557:78665]
h_7 = df[78668:91773]
h_8 = df[91777:104885]
h_9 = df[104888:]
num_vars = len(h_1)

centre = 0.5

theta = np.linspace(0, 2*np.pi, num_vars, endpoint=False)
# rotate theta such that the first axis is at the top
theta += np.pi/2

def unit_poly_verts(theta, centre ):
    """Return vertices of polygon for subplot axes.
    This polygon is circumscribed by a unit circle centered at (0.5, 0.5)
    """
    x0, y0, r = [centre ] * 3
    verts = [(r*np.cos(t) + x0, r*np.sin(t) + y0) for t in theta]
    return verts

def radar_patch(r, theta, centre ):
    """ Returns the x and y coordinates corresponding to the magnitudes of 
    each variable displayed in the radar plot
    """
    # offset from centre of circle
    offset = 0.01
    yt = (r*centre + offset) * np.sin(theta) + centre 
    xt = (r*centre + offset) * np.cos(theta) + centre 
    return xt, yt

verts = unit_poly_verts(theta, centre)
x = [v[0] for v in verts] 
y = [v[1] for v in verts] 

p = figure(title="GCMS Hazelnut Data - 501,504,509",width=1000,height=500)
text = ["" for x in range(0,num_vars+1)]
source = ColumnDataSource({'x':x + [centre ],'y':y + [1],'text':text})

p.line(x="x", y="y", source=source)

labels = LabelSet(x="x",y="y",text="text",source=source)

p.add_layout(labels)

# example factor:
f1 = np.asarray(h_1['File'],dtype=np.float32)
f2 = np.asarray(h_2['File'],dtype=np.float32)
f3 = np.asarray(h_3['File'],dtype=np.float32)
f4 = np.asarray(h_4['File'],dtype=np.float32)
f5 = np.asarray(h_5['File'],dtype=np.float32)
f6 = np.asarray(h_5['File'],dtype=np.float32)
f7 = np.asarray(h_5['File'],dtype=np.float32)
f8 = np.asarray(h_5['File'],dtype=np.float32)
f9 = np.asarray(h_5['File'],dtype=np.float32)
#xt = np.array(x)
flist = [f5,f6,f8]
names = ['505 - wepster','506 - macdonald','508 - wepster']
colors = ['blue','green','red', 'orange','purple']

# select a palette
colors = itertools.cycle(palette)    

for i, color in zip(range(len(flist)), colors):
    xt, yt = radar_patch(flist[i], theta, centre)
    p.patch(x=xt, y=yt, fill_alpha=0.15, line_width=0.0,color=color,legend=names[i])
p.legend.location='top_left'

# save the output and show the plot
output_file("plot.html")
show(p)
```
